```python
#Summary Statistics

# Mount Google Drive
from google.colab import drive
drive.mount('/content/drive')

# Import required libraries
import pandas as pd

# Load datasets from Google Drive
patients_path = "/content/drive/MyDrive/CIND820/patients.csv"
slots_path = "/content/drive/MyDrive/CIND820/slots.csv"
appointments_path = "/content/drive/MyDrive/CIND820/appointments.csv"

patients_df = pd.read_csv(patients_path)
slots_df = pd.read_csv(slots_path)
appointments_df = pd.read_csv(appointments_path)

# Define a function to clean and summarize each dataset
def clean_and_summarize(df, name):
    print(f"\n--- {name} Dataset Summary ---")

    # Remove duplicates
    initial_shape = df.shape
    df.drop_duplicates(inplace=True)
    final_shape = df.shape
    print(f"Removed {initial_shape[0] - final_shape[0]} duplicate rows.")

    # Column names and data types
    print("\nColumn Names and Data Types:")
    print(df.dtypes)

    # Missing values
    print("\nMissing Values Count:")
    print(df.isnull().sum())

    # Summary statistics
    print("\nSummary Statistics:")
    display(df.describe(include='all'))

    return df

# Clean and summarize each dataset
patients_df_clean = clean_and_summarize(patients_df, "Patients")
slots_df_clean = clean_and_summarize(slots_df, "Slots")
appointments_df_clean = clean_and_summarize(appointments_df, "Appointments")
```

```python
# Abnormal Data Identification

# Define a function to identify abnormal data
def identify_abnormal_data(df, name):
    print(f"\n--- {name} Dataset Abnormal Data Check ---")

    # Missing values
    print("\nMissing Values:")
    print(df.isnull().sum())

    # Abnormal ages
    if 'age' in df.columns:
        abnormal_ages = df[(df['age'] < 0) | (df['age'] > 120)]
        print(f"\nAbnormal Ages: {len(abnormal_ages)}")
        if not abnormal_ages.empty:
            display(abnormal_ages[['appointment_id', 'patient_id', 'age']])

    # Negative scheduling intervals
    if 'scheduling_interval' in df.columns:
        negative_sched = df[df['scheduling_interval'] < 0]
        print(f"\nNegative Scheduling Intervals: {len(negative_sched)}")
        if not negative_sched.empty:
            display(negative_sched[['appointment_id', 'scheduling_interval']])

    # Negative waiting times
    if 'waiting_time' in df.columns:
        negative_wait = df[df['waiting_time'] < 0]
        print(f"\nNegative Waiting Times: {len(negative_wait)}")
        if not negative_wait.empty:
            display(negative_wait[['appointment_id', 'waiting_time']])

# Run abnormal data checks
identify_abnormal_data(patients_df, "Patients")
identify_abnormal_data(slots_df, "Slots")
identify_abnormal_data(appointments_df, "Appointments")
```
